# Diffraction in 2D

## Theoretical background
The diffraction pattern in 2D involves the same maths as in 1D, but extending by considering the y-coordinates as well. The 2D Fast Fourier transform is well defined as well, so that is What I have used here.

```python
%matplotlib inline
import numpy.fft as fft
import numpy as np
import matplotlib.pyplot as plt

def map_to_array(function, array):
    return np.array(list(map(function, array)))

def fresnel_correction(wavelength, D):
    k = 2*np.pi/wavelength
    def correction(x, y, aperture):
        return np.exp((x**2 + y**2)/2*D)*aperture
    return correction

def circular_aperture(radius):
    radiussq = radius**2
    def _aperture(x, y):
        arr = []
        for elx in x:
            tmp = []
            for ely in y:
                if elx**2 + ely**2 < radiussq:
                    tmp.append(1.)
                else:
                    tmp.append(0.)
            arr.append(tmp)
        return np.array(arr)
    return _aperture

def square_aperture(width):
    def _aperture(x, y):
        arr = []
        for elx in x:
            tmp = []
            for ely in y:
                if abs(elx) < width and abs(ely) < width:
                    tmp.append(1.)
                else:
                    tmp.append(0.)
            arr.append(tmp)
        return np.array(arr)
    return _aperture

def slit_aperture(width):
    def _aperture(x, y):
        arr = []
        for elx in x:
            tmp = []
            for ely in y:
                if abs(ely) < width and abs(elx) < width*10:
                    tmp.append(1.)
                else:
                    tmp.append(0.)
            arr.append(tmp)
        return np.array(arr)
    return _aperture

def double_slit_aperture(width, displacement):
    aperture_internal = slit_aperture(width)
    def _aperture(x, y):
        prelim_value = aperture_internal(x, y - displacement/2)
        value = prelim_value + aperture_internal(x, y + displacement/2)
        return value
    return _aperture

def amp_2d(aperture_function, x, y, D, wavelength):
    correction = fresnel_correction(wavelength, D)
    aperture = aperture_function(x, y)
    actual_aperture = aperture#correction(x, y, aperture)
    return fft.fftshift(fft.fft2(actual_aperture))

def crop(image, x1, x2, y1, y2):
    """
    Return the cropped image at the x1, x2, y1, y2 coordinates
    """
    if x2 == -1:
        x2=image.shape[1]-1
    if y2 == -1:
        y2=image.shape[0]-1

    mask = np.zeros(image.shape)
    mask[y1:y2+1, x1:x2+1]=1
    m = mask>0
    return image[m].reshape((y2+1-y1, x2+1-x1))
```

```python
delta_x = 0.01
x_range = 5
D = 10
wavelength = 1.
num_elements = x_range/delta_x
x = map_to_array(lambda x: x - num_elements*delta_x/(2), 
                      np.arange(num_elements)*delta_x)
y = map_to_array(lambda x: x - num_elements*delta_x/(2), 
                      np.arange(num_elements)*delta_x)
aperture_function = circular_aperture(radius = 0.5)
amplitude = amp_2d(aperture_function, x, y, D, wavelength)
fig = plt.figure()
ax = fig.gca()
pattern = (abs(amplitude))**0.5
ax.imshow(aperture_function(x, y), interpolation='none', cmap='inferno')
fig = plt.figure()
ax = fig.gca()
pattern = crop(pattern, 220, 280, 220, 280) 
ax.imshow(pattern, interpolation='none', cmap='inferno')
```
